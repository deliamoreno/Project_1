### ANALÍSIS E INTERPRETACIÓN: 
##### RESPUESTA AL PUNTO  B Y C
Para eliminar los  Outliers o (las Cantidades Fuera de los Limites de la Variable) utlizo una Función para Eliminarlos Tanto para la variable  Superficie Total y Costo de esta forma Reduzco la cantidad de valores atipicos  es decir valores muy chicos o muy grandes 

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)  # para evitar la visualización de warnings

def remove_outlier(dr,n, h):    # Función que refleja la "tecnica de puntuaciónes IQR" para encontrar los valores atipicos en una variable
       
        Q1=dr.quantile (0.25) 
        Q3= dr.quantile(0.75)
        IQR= Q3 - Q1 

        lower_limit = Q1-1.5 * IQR #calculamos el outlier menor
        upper_limit =  Q3 + 1.5 * IQR #calculamos el outlier mayor

        surface_outliers = dr[~((dr < lower_limit ) | (dr > upper_limit)).any(axis=1)] # filtro que  elimina los outliers de la variable, obteniendo asi los valores dentro de los limites
        return  surface_outliers   
```
